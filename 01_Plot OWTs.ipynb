# 2 Data Access

```python
# library
import pandas as pd

# read data

rrs_norm = pd.read_csv('00_Database/01_Tables/rrsnorm_mean.csv', index_col=0)
rrs = pd.read_csv('00_Database/01_Tables/rrs_mean.csv', index_col=0)
```

# 3 Plot data

```python
# library used
import matplotlib.pyplot as plt
import seaborn as sns

# plot
f = plt.figure(figsize=(8.6,4.3))

ax = plt.subplot(1,1,1)
rrs.transpose().plot(linewidth=3,
                     color=['#02b7d7', '#0057ed', '#7cb600', '#00f625', '#188400', '#000000', '#fffb00', '#ff0004'],
                     style=['-',':', '-', '-','-',':', '-', '-'],
                     alpha=1,
                     ylim=(0,0.06),
                     ax=ax)

ax.legend(['OWT 1',
          'OWT 2',
          'OWT 3',
          'OWT 4',
          'OWT 5',
          'OWT 6',
          'OWT 7',
          'OWT 8'],
           fontsize=10)

t = ax.table([[' ',' ',' ',' ',' '],
              [' ',' ',' ',' ',' '],
              ['OWT', 'Chl-a (mg m$^{-3}$)', 'TSM (g m$^{-3}$)', 'CDOM (m$^{-1}$)', 'Z$_{sd}$ (m)'],
    ['1: Clear waters','6.04 (6.18)', '2.64 (2.94)', '0.6 (0.46)', '2.35 (1.0)'],
         ['2: Moderate turbid waters','5.28 (7.93)', '5.14 (4.1)', '1.71 (1.56)', '1.3 (0.5)'],
         ['3: Eutrophic turbid waters','33.1 (19.1)', '14.38 (10.5)', '1.76 (0.48)', '0.59 (0.34)'],
         ['4: Eutrophic clear waters','39.6 (29.7)', '6.1 (3.25)', '0.7 (0.16)', '1.76 (0.34)'],
         ['5: Hypereutrophic waters','180.4 (101)', '25.75 (11.8)', '0.98 (0.21)', '0.7 (0.33)'],
         ['6: CDOM richest waters','14.62 (9.6)', '8.2 (3.67)', '4.07 (1.4)', '1.0 (0.35)'],
         ['7: Turbid waters','8.26 (7.67)', '19.55 (13.0)', '1.8 (0.62)', '0.4 (0.15)'],
         ['8: Very turbid waters','12.13 (15.6)', '93.25 (49.1)' ,'1.18 (0.27)', '0.16 (0.06)']],
         rowLoc='left',
         colLoc='left',
         cellLoc='left',
         colWidths=[0.5, 0.3, 0.3, 0.3, 0.3],
         edges='open',
         loc='bottom',
        fontsize=16)

t.scale(1,1.4)
f.subplots_adjust(right=0.8)
ax.set_ylabel('R$_{rs}$ (sr$^{-1}$)', fontsize=10)
ax.set_xlabel('Wavelength (nm)', fontsize=10);
```

```python
f.canvas.draw()
f.savefig('00_Database/00_Figures/owts.jpeg', dpi=300, bbox_inches="tight")
```
